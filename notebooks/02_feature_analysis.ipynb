# 02 Feature Analysis

Build features and inspect distributions, correlations, and completeness.

```python
import sys, warnings
warnings.filterwarnings("ignore")
sys.path.insert(0, "..")
import numpy as np
import pandas as pd
import plotly.express as px
from config import settings
from config.logging_config import configure_logging
from src.data.fetcher import fetch_fx_data
from src.features.builder import build_features
configure_logging(level="WARNING")
```

```python
raw = fetch_fx_data(start="2018-01-01")
features = build_features(raw, persist=False)
features.shape
```

```python
block = features["EURUSD=X"]
block.describe().T.head(20)
```

## Feature correlation heatmap (EUR/USD)

```python
corr = block.dropna(axis=1, how="all").corr()
px.imshow(corr, title="EUR/USD feature correlations")
```

## Completeness per feature

```python
completeness = features.notna().mean().sort_values()
completeness.head(20)
```
