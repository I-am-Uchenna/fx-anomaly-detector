# 01 Data Exploration

Fetch live FX data, assess quality, and inspect return distributions and stationarity.

```python
import sys, warnings
warnings.filterwarnings("ignore")
sys.path.insert(0, "..")
import numpy as np
import pandas as pd
import plotly.express as px
from config import settings
from config.logging_config import configure_logging
from src.data.fetcher import fetch_fx_data
from src.features.builder import build_features
configure_logging(level="WARNING")
```

```python
raw = fetch_fx_data(start="2018-01-01")
list(raw.keys())
```

```python
eur = raw["EURUSD=X"]
eur.tail()
```

## Data quality
Completeness and gaps per pair.

```python
quality = {s: {"rows": len(df), "nan_close": int(df["close"].isna().sum())} for s, df in raw.items()}
pd.DataFrame(quality).T
```

## Return distributions

```python
rets = np.log(eur["close"]).diff().dropna()
px.histogram(rets, nbins=80, title="EUR/USD daily log returns")
```

## Stationarity (ADF on prices vs returns)

```python
from statsmodels.tsa.stattools import adfuller
print("price ADF p:", adfuller(eur["close"].dropna())[1])
print("return ADF p:", adfuller(rets)[1])
```
